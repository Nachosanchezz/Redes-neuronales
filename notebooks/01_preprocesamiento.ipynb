```python
# --- 1. Codificación OneHot ---
onehot_cols = ['estado_civil', 'tipo_empleo', 'ocupacion', 'continente_nacimiento']
columnas_a_codificar = [col for col in onehot_cols if col in df.columns]

if columnas_a_codificar:
    df = pd.get_dummies(df, columns=columnas_a_codificar, drop_first=False)

# --- 2. Conversión de columnas con comas ---
for col in df.columns:
    if df[col].dtype == 'object':
        muestras = df[col].dropna().astype(str).head(5)
        if muestras.str.contains(r'^\d+,\d+$').any():  # ejemplo: '88,6'
            df[col] = df[col].str.replace(',', '.').astype(float)

# --- 3. Mostrar columnas no convertibles (tipo texto real, como 'sexo') ---
no_convertibles = df.select_dtypes(include='object').columns.tolist()
if no_convertibles:
    print("Estas columnas aún tienen texto y no se pueden convertir:", no_convertibles)

# --- 4. Convertir el resto a float ---
df[df.select_dtypes(exclude='object').columns] = df.select_dtypes(exclude='object').astype(float)
```

```python
df['actividad_fisica'].unique()
```

```python
ordinal_cols = ['nivel_educativo', 'actividad_fisica']

orden_nivel_educativo = ['Primaria', 'Secundaria', 'Universitaria', 'Posgrado']
orden_actividad_fisica = ['Baja', 'Moderada', 'Alta']

oe = OrdinalEncoder(categories=[orden_nivel_educativo, orden_actividad_fisica])
```

```python
df[ordinal_cols] = oe.fit_transform(df[ordinal_cols])
```

```python
# Codificar sexo como binario
df['sexo'] = df['sexo'].map({'F': 0, 'M': 1})
```

```python
# --- 2. Detectar columnas continuas (numéricas con más de 2 valores distintos) ---
columnas_continuas = [col for col in df.select_dtypes(include=['float64', 'int64']).columns
                      if df[col].nunique() > 2]

# --- 3. Normalizar columnas continuas ---
scaler = StandardScaler()
df[columnas_continuas] = scaler.fit_transform(df[columnas_continuas])
```
